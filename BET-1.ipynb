```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

data_ads = pd.read_excel('BET.xlsx')

plt.rcParams['font.sans-serif'] = ['Times New Roman']
plt.rcParams['mathtext.fontset'] = 'cm'
plt.rcParams['axes.unicode_minus'] = False

sample_names = ['400°C', '500°C', '600°C', '700°C', '800°C',]
y_offsets = [-10, 50, 100, 150, 189]  
colors = ['#2ca02c', '#1f77b4', '#ff7f0e', '#d62728', '#9467bd'] 
markers = ['o', 's', '^', 'D', 'v']
fig, ax = plt.subplots(figsize=(8, 6))
for i in range(5):
    offset = y_offsets[i]
    color = colors[i]
    marker = markers[i]
    name = sample_names[i]

    base = 4 * i  

    x_ads = data_ads.iloc[:, base]
    y_ads = data_ads.iloc[:, base + 1] + offset
    ax.plot(x_ads, y_ads, marker=marker, markersize=5, linestyle='-', linewidth=1.8,
            color=color, alpha=0.9, label=name)

    x_des = data_ads.iloc[:, base + 2]
    y_des = data_ads.iloc[:, base + 3] + offset
    ax.plot(x_des, y_des, marker=marker, markersize=5, linestyle='-', linewidth=1.8,
            color=color, alpha=0.9)


ax.set_title('', fontsize=12)
ax.set_xlabel('Relative Pressure ($P/P_0$)', fontsize=18, labelpad=10,fontweight='bold')
ax.set_ylabel('Volume (cm$^3$ g$^{-1}$)', fontsize=18, labelpad=10,fontweight='bold')

ax.tick_params(axis='both', which='major', labelsize=20)
for label in ax.get_xticklabels() + ax.get_yticklabels():
    label.set_fontweight('bold')

legend = ax.legend(loc='upper right', bbox_to_anchor=(0.19, 1.0), fontsize=14,
                   frameon=True, fancybox=True, shadow=False, ncol=1)

for text in legend.get_texts():
    text.set_fontweight('bold')

for spine in ax.spines.values():
    spine.set_linewidth(2)
ax.grid(False)
ax.set_xlim(0, 1.0)
ax.set_ylim(bottom=0)
ax.set_yticks([])
plt.tight_layout()
plt.savefig('adsorption_isotherm_shifted.png', dpi=1000, bbox_inches='tight')
plt.show()
```
